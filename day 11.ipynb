```python
#prime integers - reverse order
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

n= int(input("enter number of integers"))
numbers = list(map(int,input("enter the the numbers").split()))
prime_numbers = [num for num in numbers if is_prime(num)]
prime_numbers.reverse()
print(prime_numbers)
```

```python
#first four factors
numbers= int(input("enter number of integers"))
factors = []
for i in range(1,numbers+1):
    if numbers%i == 0:
        factors.append(i)
    if len(factors) == 4:
        break
print(factors)
```
